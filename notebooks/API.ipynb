```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from tqdm import tqdm
import time
```

```python
def holidays_api_call(props, url = "https://holidays.abstractapi.com/v1/"):
    country = props['country']
    year = props['year']
    month = props['month']
    day = props['day']

    response = requests.get(url, params={
        "api_key": "5a266c12bafb4680a128f4f9c44f450e",
        "country": country,
        "year": year,
        "month": month,
        "day": day
    })
    if response.status_code == 200:
        return response.json()
    else:
        raise Exception(f"API call failed with status code {response.status_code}")
```

```python
df = pd.read_csv('final_df.csv')
```

```python
all_dates = df['eventdate'].value_counts().sort_index().keys().tolist()
```

```python
all_dates_dict = {}
all_dates_country = {}
```

```python
curr_country = "RU"
```

```python
for date in tqdm(all_dates):
    if (date, curr_country) not in all_dates_country:
        curr_date = pd.to_datetime(date, format='%Y-%m-%d')
        curr_year = curr_date.year
        curr_month = curr_date.month if curr_date.month >= 10 else f"0{curr_date.month}"
        curr_day = curr_date.day if curr_date.day >= 10 else f"0{curr_date.day}"
        try:
            all_dates_country[(date, curr_country)] = holidays_api_call(props = {
                'year': str(curr_year),
                'month': str(curr_month),
                'day': str(curr_day),
                'country': curr_country
            })
            time.sleep(1)  
        except Exception as e:
            print(f"Error for date {date}: {e}")
```
