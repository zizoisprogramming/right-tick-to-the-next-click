```python
pytrends = TrendReq(hl="en-US", tz=360)
```

```python
def get_trend_score(keyword, publish_date):
    """
    Get the weekly search interest score for a keyword
    around the video's publish date.
    """
    try:
        # Build a 4-week window around publish date
        start = pd.to_datetime(publish_date) - pd.Timedelta(weeks=1)
        end   = pd.to_datetime(publish_date) + pd.Timedelta(weeks=1)
        timeframe = f"{start.date()} {end.date()}"

        pytrends.build_payload(
            kw_list   = [keyword],
            timeframe = timeframe,
            gprop     = "youtube"   # YouTube-specific trends
        )

        df = pytrends.interest_over_time()

        if df.empty:
            return 0

        # Return the score closest to publish date
        return df[keyword] #.mean()

    except Exception as e:
        print(f"Error for {keyword}: {e}")
        return None

    finally:
        time.sleep(1)  # avoid rate limiting
```

```python
get_trend_score("Trump Presidency", '2017-11-13T11:17:36Z')
```
